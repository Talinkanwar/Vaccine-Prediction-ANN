```python
for i in data:
  if((data[i].dtype == "float64") and (data[i].isna().sum() > 0)):
    print(i)
    data[i].fillna(data[i].mode()[0] , inplace = True)
```

```python
for i in data:
  if(data[i].dtype =="object" and data[i].isna().sum()>0):
    data[i].fillna(data[i].mode()[0] , inplace = True)
    print(i)
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for i in data:
  if(data[i].dtype == "object"):
    data[i] = le.fit_transform(data[i])
```

```python
data.head()
```

```python
from sklearn.feature_selection import chi2
```

```python
x1 = data.drop("h1n1_vaccine" , axis = 1)
y1 = data["h1n1_vaccine"]
```

```python
score = chi2(x1,y1)
```

```python
score
```

```python
p_value = pd.Series(score[1] , index = x1.columns)
```

```python
p_value
```

```python
temp = []
for i in p_value.index:
    if(p_value[i] <= 0.05):
        print(i, " : Not accepted")
    else:
        print(i, " : accepted")
        temp.append(i)
```

```python
temp
```

```python
data.drop(temp ,axis = 1 , inplace = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x = data.drop("h1n1_vaccine" , axis = 1)
y = data["h1n1_vaccine"]
```

```python
new_x = sc.fit_transform(x)
```

```python
x_train , x_test , y_train , y_test = train_test_split(new_x , y , test_size = 0.25 , random_state = 40)
```

```python
x_train.shape
```

```python
import keras
```

```python
import keras.models as km
```

```python
from keras import initializers , activations , regularizers , constraints
```

```python
from keras.layers import Dense , Activation
```

```python
from  keras import Sequential
```

```python
from keras.metrics import AUC
```

```python
sq = Sequential()
```

```python
x_train.shape
```

```python
sq.add(Dense(units = 27 , kernel_initializer = "uniform" , activation ="relu" , input_dim = 27 ))
sq.add(Dense(units = 24 , kernel_initializer = "uniform" , activation ="relu" ))
sq.add(Dense(units = 21 , kernel_initializer = "uniform" , activation ="relu" ))
sq.add(Dense(units = 18 , kernel_initializer = "uniform" , activation ="relu" ))
sq.add(Dense(units = 15 , kernel_initializer = "uniform" , activation ="relu" ))
sq.add(Dense(units = 12 , kernel_initializer = "uniform" , activation ="relu" ))
sq.add(Dense(units = 9 , kernel_initializer = "uniform" , activation ="relu" ))
sq.add(Dense(units = 6 , kernel_initializer = "uniform" , activation ="relu" ))
sq.add(Dense(units = 3 , kernel_initializer = "uniform" , activation ="relu" ))
sq.add(Dense(units = 2 , kernel_initializer = "uniform" , activation ="relu" ))
sq.add(Dense(units = 1 , kernel_initializer = "uniform" , activation ="sigmoid" ))
```

```python
sq.compile(optimizer = "adam" , loss = "binary_crossentropy" , metrics = ["accuracy" ,"AUC"])
```

```python
sq.fit(x_train , y_train , batch_size = 100 , epochs = 1000)
```
